# Threat Hunting with CIRCL (open-source) Tools
Welcome!  
In this workshop, you'll learn to:
- Ingest and correlate IOCs in **MISP**
- Enrich data with **PassiveDNS**, **PassiveSSL**, and **PassiveSSH**
- Geolocalize IPs with **mmdb-server**.
- Identify malware with **Hashlookup** and CVEs via  **Vulnerability Lookup**
- Investigate Ransomware groups with **Ransomlook.io** and **OnionLookup**
- Analyze files in a sandbox using **Pandora**
- Visualize phishing or malicious domains with **Lookyloo**

Let's begin!
